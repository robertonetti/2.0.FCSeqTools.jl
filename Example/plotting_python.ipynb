```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
single_dkl = open("single_dkl.txt","r")
total_dkl = open("total_dkl.txt", "r")

lines = single_dkl.readlines()[2 : ]
lines2 = total_dkl.readlines()[2 : ]  
```

```python
x = []
for line in lines[0:-1]:
    line = line[1:-2]
    line = line.split(", ")
    for i in range(0,len(line)):
        line[i] = float(line[i])
    x.append(line)

last_line = lines[-1][1:-1].split(", ")
for i in range(0,len(last_line)):
    last_line[i] = float(last_line[i])
x.append(last_line)

    


y = []
for line in lines2:
    y.append(float(line))
   
```

```python
len(x)
```

```python
sum_x = []
for i in x:
    sum_x.append(sum(i))
sum_x
```

```python
print(len(sum_x))
```

```python
plt.plot(y, sum_x, "*")
plt.grid("True")
plt.xlabel("full_DKL")
plt.ylabel("sum of singles")

```

```python

tot = 0
for i in x:
    fi = np.sort(i) / sum(i)
    tot += fi
tot /= len(x)


plt.bar(range(0,25),np.sort(x[97]), 1)
plt.show()



plt.bar(range(0,25), tot, 1)
```
